```python
import random

def add_rand(txt:dict):
    res = ""
    for key, item in txt.items():
        res += f"{key} "*random.randint(1, item)
    return res
```

```python
my_skills = {"Python":9, "Dusk":1, "Data_Wrangling":5, "SciKit_Learn":5, "Dusk_ML":1, "Visualization":3,
            "R":1, "julia":1, "Java":1, "SQL":3, "PyTorch":2, "KNIME":1, "Power_BI":4, "Jupyter_lab":4, "Data":6,
            "Google_Colabs":2, "NumPy":5, "Pandas":5, "Matplotlib":2, "TensorFlow":3, "Artificial_Intelligence":4,
            "Machine_Learning":4, "Visual_Studio_Code":2, "Data_Science":2}
```

```python
data = add_rand(my_skills)
# add other words
data += " Supervised Unsupervised Overfitted Underfitted Classification Regression Hadoop Spark Training Testing"
data += " DataFrame Array Data_Lake SVM Neuronal_Networks Gaussian KMeans Mean_Shift DBSCAN F1_Score OPTICS"
data += " Azure Language_Models Word2Vec Bag_of_Words AI KÃ¼nstliche_Intelligenz KI n_gram RNN CNN NN AlexNet"
data += " Devset 10_Fold_Cross_Validation Random_Forest Decision_Tree Eager_Learner Lazy_Learner Clustering"
data += " MapReduce Gradient Optimization Coffee 404 Error Profiling HS_Offenburg Curse_of_Dimensionality Feed_Forward Naive_Bayes"
data += " Soft_Voting Cosin Sine Similarity Datapoints Abitrary Noise Quality Quantity Outlier Outlier_Detection"
data += " Hyperparameter Finetuning Accuracy Intellegence Future Possibilities Feature_Engineering Feature"
data += " Dimension CRISP_DM Computer_Vision Data_Scraping Data_Collection Data Learning Algorithms Transformers"
data += " NLP Language_Models LSTM"
```

```python
# text analysis
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(data)
```

```python
# create wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(collocations=False, width=2000, height=2000, background_color = "white").generate(doc.text)

plt.figure(figsize = (10,10))
plt.axis("off")
plt.imshow(wordcloud, interpolation = "bilinear")
```
